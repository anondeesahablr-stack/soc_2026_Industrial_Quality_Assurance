Lists:
They are ordered, mutable datatypes that allows multiple datatypes and duplicate elements.
To access element, do so using its index. (indices start at 0)
Negative indices start from -1 to -len, which goes in revrese order
in - can be used to iterate through list elements, to check if element is inside list

List methods:
len(mylist)
mylist.append("token")
mylist.insert(1,"toss")
mylist.pop() #removes last element 
mylist.remove("token") #returns valueError is not in list
mylist.clear() #removes all elements in list
mylist.reverse() 
mylist.sort()
sorted_list = sorted(mylist) #reverse=true
newlist = [0]*5
newlist1 = mylist + sorted_list

Slicing:
newlist = mylist[1:5:2] #all numbers from index 1 to 4 step by 2 - 1,3
reverse = mylist[::-1]

With the direct assignment way of copying a list to another, both refer to same address so changes to one is reflected on the other. Instead use - newlist = mylist.copy()
Or use newlist = list(mylist)
Or use newlist = mylist[:]

```python
#list comprehension
mylist=[1,2,3,4,5,6]
b=[i*i for i in mylist]
print(mylist)
print(b)
```

Tuples:
They are ordered, immutable, allows duplicate elements
Parentheses are optional, comma is compulsory even if there's only one element
mytuple = tuple([1,2,3,4])
mytuple[0]="Hello" - ERROR as tuple is immutable
mytuple.count("p") - returns number of times p is present in mytuple
mytuple.index("p") - first instance of the element is returned, error if element is not present

Unpacking tuples:

rollno, name, city = mytuple
numbers = (1,2,3,4)

```python
#working with tuples can be more efficient than working with lists
import sys
import timeit
my_list = [0,1,2,"hello",True]
my_tuple = (0,1,2,"hello",True)
print(sys.getsizeof(my_list), "bytes")
print(sys.getsizeof(my_tuple), "bytes")
print(timeit.timeit(stmt="[1,2,3,4,5]",number=1000000))
print(timeit.timeit(stmt="(1,2,3,4,5)",number=1000000))
```

```python
odds = {1,3,5,7,9}
evens = {0,2,4,6,8}
primes = {2,3,5,7}
u = odds.union(evens)
print(u)
i = odds.intersection(primes)
print(i)
```

```python
mystring = "how,are,you,doing"
my_list = mystring.split(',')
print(my_list)
newstr = ' '.join(my_list)
print(newstr)
```

```python
#formatting
var = 3.127
name = "Tom"
# f - float, d - int, s - string .2 - number of decimal points
mystr = "the variable is %.2f" % var
print(mystr)

mystr1 = "the variable is {}".format(var)
print(mystr1)
mystr2= "the variable is {:.2f} and {}".format(var,name)
print(mystr2)

mystr3 = f"the variable is {var} and {name}"
print(mystr3)
```

Collections:
Counter, namedtuple, OrderedDict, defaultdict, deque
Counter - dictionary so you can access keys and values with .keys(), .values(), .items()
namedtuple - like struct
OrderedDict - remembers the order in which items were inserted
defaultdict - can set default data type during creation
deque - can add elements to both sides using .appendleft(), .extendleft()

```python
from collections import Counter
a = "aaaabbbcc"
my_counter = Counter(a)
print(my_counter)
print(my_counter.most_common(2)[1][0])
print(list(my_counter.elements()))
```

```python
from collections import namedtuple
Point = namedtuple('Point', 'x,y')
pt = Point(1,-4)
print(pt)
print(pt.x)
```
