PANDAS:
Pandas is an open-source Python library designed for data manipulation and analysis. 
Two main types - DataFrame, Series (row/column)
DataFrame - A 2D labeled data structure (like a full spreadsheet or SQL table)
Series - A 1D labeled array (like a column)
